```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import Math
```

```python
baris = np.array([2,3])
kolom = np.array([[1],[4]])
```

```python
print(kolom[0])
```

```python
print(baris)
print(kolom)
```

```python
display(Math(sym.latex(baris)))
display(Math(sym.latex(kolom)))
```

```python
display(Math(sym.latex(sym.sympify(baris))))
display(Math(sym.latex(sym.sympify(kolom))))
```

```python
display((sym.sympify(baris)))
display((sym.sympify(kolom)))
```

```python
plt.plot([0,baris[0]])
# plt.plot([[0,kolom[1]]])

plt.axis('square')
plt.axis([-5,5, -5,5])
plt.show()
```

```python
plt.plot([0,baris[0]],[0,baris[1]])
```

```python
plt.plot([0,baris[0]],[0,baris[1]], 'r', label = 'vektor baris')
# plt.plot([0,kolom[0]],[0,kolom[1]], 'b', label = 'vektor kolom')

plt.axis('square')
plt.axis([-5,5, -5,5])
plt.grid()
plt.legend()
plt.show()
```
